```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
print(housing.feature_names)
```

* housing.data are the x values
* housing.target are the y values

```python
# Import data
from sklearn.model_selection import train_test_split
X_train_full, X_test, y_train_full, y_test = train_test_split(housing.data, housing.target, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full, random_state=42)
```

```python
# Pre-process data (i.e. normalize it)
# Standard Scalar will essentially make it so each X_field has a mean of 0 and varience of 1
# Needed when each of the colums (i.e. input data) has different ranges
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)
```

```python
# Set random seeds so we get same results each time model is run
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# 8 Columns for 11610 rows
X_train.shape
```

```python
# Create structure for regression neural network
model = keras.Sequential([
    # input_shape is the number of inputs for X_data e.g input_shape = X_train.shape[1:]
    keras.layers.Dense(30, activation="relu", input_shape=[8]),
    keras.layers.Dense(30, activation="relu"),
    keras.layers.Dense(1) # Regression since output is a single value
])
```

```python
model.summary()
```

```python
# Compile model
# we use a loss funciton of mean_squared_error since we are doing a linear regrssion model
model.compile(
    loss="mean_squared_error",
    optimizer=keras.optimizers.SGD(lr=1e-3), # lr means learning rate, default is 0.01
    metrics=["mae"] # in regression use mean absolute error (square of differenece between predicted and actual value)
                    # We will use this to judge how well our model is doing
)
```

```python
# Train model
model_history = model.fit(X_train, y_train, epochs=20, validation_data=(X_valid, y_valid))
# you can run this multiple times - since keras remembers the biases from previous runs
```

```python
# Test the model
mae_test = model.evaluate(X_test, y_test)
```

```python
# Results:
# FIRST RUN: loss: 0.3033 - mae: 0.4493
# SECOND RUN: loss: 0.2515 - mae: 0.4257
```

```python
model_history.history
```

```python
# Plot model history
pd.DataFrame(model_history.history).plot(figsize=(8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```
