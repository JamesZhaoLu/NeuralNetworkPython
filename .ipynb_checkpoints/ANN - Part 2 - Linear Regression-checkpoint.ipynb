```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
print(housing.feature_names)
```

* housing.data are the x values
* housing.target are the y values

```python
# Import data
from sklearn.model_selection import train_test_split
X_train_full, X_test, y_train_full, y_test = train_test_split(housing.data, housing.target, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full, random_state=42)
```

```python
# Pre-process data (i.e. normalize it)
# Standard Scalar will essentially make it so each X_field has a mean of 0 and varience of 1
# Needed when each of the colums (i.e. input data) has different ranges
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)
```

```python
# Set random seeds so we get same results each time model is run
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# 8 Columns for 11610 rows
X_train.shape
```

```python
# Create structure for regression neural network
model = keras.Sequential([
    # input_shape is the number of inputs for X_data e.g input_shape = X_train.shape[1:]
    keras.layers.Dense(30, activation="relu", input_shape=[8]),
    keras.layers.Dense(30, activation="relu"),
    keras.layers.Dense(1) # Regression since output is a single value
])
```

```python
model.summary()
```

```python
# Compile model
# we use a loss funciton of mean_squared_error since we are doing a linear regrssion model
model.compile(
    loss="mean_squared_error",
    optimizer=keras.optimizers.SGD(lr=1e-3), # lr means learning rate, default is 0.01
    metrics=["mae"] # in regression use mean absolute error (square of differenece between predicted and actual value)
                    # We will use this to judge how well our model is doing
)
```

```python
# Train model
model_history = model.fit(X_train, y_train, epochs=20, validation_data=(X_valid, y_valid))
# you can run this multiple times - since keras remembers the biases from previous runs
```

```python
# Test the model
mae_test = model.evaluate(X_test, y_test)
```

```python
# Results:
# FIRST RUN: loss: 0.3033 - mae: 0.4493
# SECOND RUN: loss: 0.2515 - mae: 0.4257
```

```python
model_history.history
```

```python
# Plot model history
pd.DataFrame(model_history.history).plot(figsize=(8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

```python
# Predict new values
X_new = X_test[:3]
```

```python
y_pred = model.predict(X_new)
print(y_pred)
print(y_test[:3])
```

## Functional API Models

```python
del model
```

```python
keras.backend.clear_session()
```

```python
# Rebuild the model using a functional model 
# In which the input layer is also fed to the output layer
# The reason this "wide" layer is better is because in sequetial models some patterns of the data
#   may be distorted by the simple layer to layer model which is not good

# We first create the layers and link them together
# It's called a functional API cause linking them together is like calling a function -> where input param
#   of the function is the previous layer
input_ = keras.layers.Input(shape=X_train.shape[1:])
hidden1 = keras.layers.Dense(30, activation="relu")(input_)
hidden2 = keras.layers.Dense(30, activation="relu")(hidden1)
concat = keras.layers.concatenate([input_, hidden2]) # concantenate layer just merges the two layers output
output = keras.layers.Dense(1)(concat) # no activation function since this is the output layer

# Then combine linked layers into a model
model = keras.models.Model(inputs=[input_], outputs=[output])
```

```python
# Print out model
model.summary()
```

```python
# Compile the model with same parameters as the Sequential example from above
model.compile(
    loss="mean_squared_error",
    optimizer=keras.optimizers.SGD(lr=1e-3),
    metrics=["mae"]
)
```

```python
model_history = model.fit(X_train, y_train, epochs=40, validation_data=(X_valid, y_valid))
```

```python
model_history.history
```

```python
# test how good this was

mae_test = model.evaluate(X_test, y_test)
```

```python
# Values of 
# First run (40 epochs): Loss 0.3064, mae 0.4293
# Compared to the previous normal Sequential NN model it was not as good but there are situations where Deep and Wide
#   models perform better
```

```python
model_history.history
```

```python
# Plot functional model history
pd.DataFrame(model_history.history).plot(figsize=(8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

```python
# Predict new values
X_new = X_test[:3]
y_pred = model.predict(X_new)
print(y_pred)
print(y_test[:3])
```

###### Note you can only save a model once it's done training OR use checkpoints to save your model after each epoch

```python
# Let's try this our
keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# Now let's reuse our Sequential model
# Create structure for regression neural network
model = keras.Sequential([
    # input_shape is the number of inputs for X_data e.g input_shape = X_train.shape[1:]
    keras.layers.Dense(30, activation="relu", input_shape=[8]),
    keras.layers.Dense(30, activation="relu"),
    keras.layers.Dense(1) # Regression since output is a single value
])
```

```python
# Compile model
# we use a loss funciton of mean_squared_error since we are doing a linear regrssion model
model.compile(
    loss="mean_squared_error",
    optimizer=keras.optimizers.SGD(lr=1e-3), # lr means learning rate, default is 0.01
    metrics=["mae"] # in regression use mean absolute error (square of differenece between predicted and actual value)
                    # We will use this to judge how well our model is doing
)
```

```python
checkpoint_cb = keras.callbacks.ModelCheckpoint("Model-{epoch:02d}.h5") # the name for each of the checkpoints
# You need to create a new name for each epoch -> so we use the {epoch:02d} variable this will be replaced by 01,02...
# epoch is the variable name 
# :02d means you just want the first 2 digits of the epoch number 01-99
```

```python
# now we retrain the model with checkpoints
history = model.fit(X_train, y_train, epochs=10,
                    validation_data=(X_valid,y_valid),
                    callbacks=[checkpoint_cb]) # this is the new parameter to save the model after each epoch
```

```python
# Now we can load a model at any epoch
del model
keras.backend.clear_session()
```

```python
model = keras.models.load_model("Model-10.h5")
```

```python
# Now we can use this model to test anytime
mse_test = model.evaluate(X_test, y_test)
```

### Best Model

```python
# However, what if it isn't the last epoch's model isn't the best - there is a way to only fetch the best model
```

```python
del model
keras.backend.clear_session()
```

```python
# Now let's reuse our Sequential model
# Create structure for regression neural network
model = keras.Sequential([
    # input_shape is the number of inputs for X_data e.g input_shape = X_train.shape[1:]
    keras.layers.Dense(30, activation="relu", input_shape=[8]),
    keras.layers.Dense(30, activation="relu"),
    keras.layers.Dense(1) # Regression since output is a single value
])
```

```python
# Compile model
# we use a loss funciton of mean_squared_error since we are doing a linear regrssion model
model.compile(
    loss="mean_squared_error",
    optimizer=keras.optimizers.SGD(lr=1e-3), # lr means learning rate, default is 0.01
    metrics=["mae"] # in regression use mean absolute error (square of differenece between predicted and actual value)
                    # We will use this to judge how well our model is doing
)
```

```python
checkpoint_cb = keras.callbacks.ModelCheckpoint("Best_Model.h5", save_best_only=True) 
# The save_best_only parameter makes it so we only save the best model at the end of all the epochs in the name
#   Best_Model.h5
```

```python
# now we retrain the model with checkpoints
history = model.fit(X_train, y_train, epochs=10,
                    validation_data=(X_valid,y_valid),
                    callbacks=[checkpoint_cb]) # this is the new parameter to save the model after each epoch
```

```python
# We only have 1 file named Best_Model.h5 in the directory now
```

```python
# We can also use best_model in tandem with another feature - EarlyStopping
# Early Stopping stops our model from training after it shows no improvement over a patience (parameter) number
#   of epochs. Then in tandem with best model only we will get the best model so far
del model
keras.backend.clear_session()
```
